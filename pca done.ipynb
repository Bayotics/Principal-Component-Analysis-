```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('rent_lagos_updated.csv')
df
```

```python
df = df.drop(columns = ['housing_type', 'bathrooms', 'guest_toilet', 'parking_space', 'district', 'address', 'local government'])
df
```

```python
df['budget'] = np.random.randint(80000, 5000000, df.shape[0])
df['cooking'] = np.random.randint(1, 5, df.shape[0])
df['leavehome'] = np.random.randint(1, 4, df.shape[0])
df['religion'] = np.random.randint(1, 3, df.shape[0])
df['sound'] = np.random.randint(1, 4, df.shape[0])
df['workout'] = np.random.randint(1,4, df.shape[0])
df
```

```python
df = df.round({'budget':-4})
df
```

```python
newDataset = df.drop(columns = ['rent_per_annum', 'bedrooms', 'districtID'])
newDataset
```

```python
def immigrants_segment(value):
    if 80000 <= value < 300000:
        return 1
    elif 300000 <= value < 600000:
        return 2
    elif 600000 <= value < 1000000:
        return 3
    elif value > 1000000:
        return 4

newDataset['immigrants_segment'] = newDataset['budget'].map(immigrants_segment)
newDataset
```

```python
newDataset.fillna('')
```

```python
n = 100
newDataset = newDataset.iloc[:n] 
newDataset
```

```python
x = newDataset.iloc[:, 0:6].values
y = newDataset.iloc[:, 6].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green', 'blue')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend
```
